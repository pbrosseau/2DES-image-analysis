```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import gaussian_filter
import cv2
```

# Monochrome Gaussian Filter

```python
im = np.zeros((9,9,1)).astype(np.uint8)
im[4,4,0] = 255

im_conv_3x3 = cv2.GaussianBlur(im, ksize=(3,3), sigmaX=0, borderType = cv2.BORDER_REFLECT)
im_conv_5x5 = cv2.GaussianBlur(im, ksize=(5,5), sigmaX=0, borderType = cv2.BORDER_REFLECT)
```

```python
plt.figure(figsize=(2,2),dpi=200)
plt.imshow(im,cmap="gray",vmin=0,vmax=255)
plt.xticks([0,2,4,6,8])
plt.title("Original")

plt.figure(figsize=(2,2),dpi=200)
plt.imshow(im_conv_3x3,cmap="gray",vmin=0,vmax=255)
plt.xticks([0,2,4,6,8])
plt.title("Kernel 3x3")

plt.figure(figsize=(2,2),dpi=200)
plt.imshow(im_conv_5x5,cmap="gray",vmin=0,vmax=255)
plt.xticks([0,2,4,6,8])
plt.title("Kernel 5x5")
```

# Color Gaussian Filter

```python
im = np.zeros((9,9,3)).astype(np.uint8)
im[:,:,2] = 255
im[4,4,2] = 0
im[4,4,1] = 255

im_conv1 = cv2.GaussianBlur(im, ksize=(3,3), sigmaX=0, borderType = cv2.BORDER_REFLECT)
im_conv2 = cv2.GaussianBlur(im, ksize=(5,5), sigmaX=0, borderType = cv2.BORDER_REFLECT)
```

```python
plt.figure(figsize=(2,2),dpi=200)
plt.imshow(im)
plt.xticks([0,2,4,6,8])
plt.title("Original")

plt.figure(figsize=(2,2),dpi=200)
plt.imshow(im_conv1)
plt.xticks([0,2,4,6,8])
plt.title("Kernel 3x3")

plt.figure(figsize=(2,2),dpi=200)
plt.imshow(im_conv2)
plt.xticks([0,2,4,6,8])
plt.title("Kernel 5x5")

plt.figure(figsize=(2,1),dpi=200)
plt.plot(im[4,:,1]/255,marker="o",label="Green",c="g")
plt.plot(im[4,:,2]/255,marker="o",label="Blue",c="b")
plt.xlabel("Pixel X-position")
plt.ylabel("Intensity")
# plt.legend()
# plt.yticks([0,63,127,191,255])
plt.xticks([0,2,4,6,8])

plt.figure(figsize=(2,1),dpi=200)
plt.plot(im_conv1[4,:,1]/255,marker="o",label="Green",c="g")
plt.plot(im_conv1[4,:,2]/255,marker="o",label="Blue",c="b")
plt.xlabel("Pixel X-position")
# plt.ylabel("Intensity")
# plt.legend()
# plt.yticks([0,63,127,191,255])
plt.xticks([0,2,4,6,8])

plt.figure(figsize=(2,1),dpi=200)
plt.plot(im_conv2[4,:,1]/255,marker="o",label="Green",c="g")
plt.plot(im_conv2[4,:,2]/255,marker="o",label="Blue",c="b")
plt.xlabel("Pixel X-position")
# plt.ylabel("Intensity")
# plt.legend()
# plt.yticks([0,63,127,191,255])
plt.xticks([0,2,4,6,8])
```
